```python
from PIL import Image
import numpy as np
```

```python
img = Image.open('pic.jpg')
x = np.array(img, dtype=np.float32)
img
```

```python
picbw = x.copy()
picbw = picbw[:,:,0]
picbw = Image.fromarray(np.asarray(picbw, dtype=np.uint8))
picbw
```

```python
x_red=x.copy()
x_red[:,:,1]=0
x_red[:,:,2]=0
img_red=Image.fromarray(np.asarray(x_red, dtype=np.uint8))
img_red
```

```python
def downrank(U, S, Vh, r):
    y_r= np.float16(U[:,:r]).dot(np.float16(np.diag(S[:r]))).dot(np.float16(Vh[:r, :]))
    y_r[y_r > 255] = 255
    y_r[y_r < 0] = 0
    img_y_r = np.asarray(y_r, dtype=np.uint8)
    return img_y_r
```

```python
U, S, Vh = np.linalg.svd(x[:,:,0], full_matrices=False)
img_red = downrank(U, S, Vh, 1)
U, S, Vh = np.linalg.svd(x[:,:,1], full_matrices=False)
img_blue = downrank(U, S, Vh, 1)
U, S, Vh = np.linalg.svd(x[:,:,2], full_matrices=False)
img_green = downrank(U, S, Vh, 1)
y = np.concatenate((img_red[:, :, np.newaxis], img_blue[:, :, np.newaxis], img_green[:, :, np.newaxis]), axis=-1)
Image.fromarray(np.asarray(y, dtype=np.uint8)).save('pic1.jpg')
Image.fromarray(np.asarray(y, dtype=np.uint8))
```
